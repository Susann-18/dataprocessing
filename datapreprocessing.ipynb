```python
import pandas as pd

# Sample dataset
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'Jake'],
    'Age': [25, 35, None, 28, 40],
    'Gender': ['M', 'F', 'M', 'F', 'M'],
    'Income': [50000, 60000, 70000, None, 55000]
}
print(data)
# Create a DataFrame
df = pd.DataFrame(data)
df
# # Data Cleaning: Handling Missing Values
df['Age'].fillna(df['Age'].mean(), inplace=True)  # Fill missing Age values with mean
df
df['Income'].fillna(df['Income'].median(), inplace=True)  # Fill missing Income values with median
df
# # Data Transformation: Convert categorical Gender variable to numerical
df['Gender'] = df['Gender'].map({'M': 0, 'F': 1})
df

```

```python
from sklearn.preprocessing import LabelEncoder

# Sample dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Label encode the 'Gender' column
df['Gender_LabelEncoded'] = label_encoder.fit_transform(df['Gender'])

# Display original and label encoded data
print("Original Data:")
print(df['Gender'])

print("\nLabel Encoded Data:")
print(df['Gender_LabelEncoded'])
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample dataset
data = {
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue']
}

# Create a DataFrame
df = pd.DataFrame(data)

# One-hot encoding using pandas
df_encoded = pd.get_dummies(df, columns=['Color'], prefix=['Color'])

# Display original and one-hot encoded data using pandas
print("Original Data:")
print(df)

print("\nOne-Hot Encoded Data:")
print(df_encoded)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample dataset
data = {
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue']
}

# Create a DataFrame
df = pd.DataFrame(data)

# One-hot encoding using scikit-learn
encoder = OneHotEncoder(sparse=False)
encoded_array = encoder.fit_transform(df[['Color']])

# Manually create column names for one-hot encoded features
column_names = encoder.get_feature_names_out(['Color'])

# Convert encoded array into DataFrame
df_encoded_sklearn = pd.DataFrame(encoded_array, columns=column_names)

# Display one-hot encoded data using scikit-learn
print("\nOne-Hot Encoded Data (Using scikit-learn):")
print(df_encoded_sklearn)
```
