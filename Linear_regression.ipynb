```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
np.random.seed(42)
x=2*np.random.rand(100,1)
y=4+3*x+np.random.rand(100,1)
x

```

```python
plt.scatter(x,y,alpha=0.75)
plt.xlabel('Independent varible (X)')
plt.ylabel('Dependent varible (y)')
plt.title('linear regression')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(34)
x1=2*np.random.rand(100,1)
x2=3*np.random.rand(100,1)
y=4+2*x1+3*x2+np.random.rand(100,1)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x1,x2,y,marker='o')
ax.set_xlabel('independent varible 1(x1)')
ax.set_ylabel('independent varible 2(x2)')
ax.set_zlabel('dependent varible (y)')
plt.title('Multilinear Regression ')
plt.show()
```
